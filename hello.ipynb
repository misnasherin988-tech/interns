```python
df['average_score'] = df[['Previous_Scores', 'Exam_Score']].mean(axis=1)
print(df.head())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print(df.columns)
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
for col in df.select_dtypes(include='number'):
    df[col] = df[col].fillna(df[col].median())
```

```python
for col in df.select_dtypes(include='object'):
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
if 'previous_score' in df.columns and 'exam_score' in df.columns:
    df['average_score'] = df[['previous_score', 'exam_score']].mean(axis=1)
```

```python
if 'hours_studied' in df.columns and 'attendance' in df.columns:
    df['study_efficiency'] = df['hours_studied'] * df['attendance']
```

```python
df = pd.get_dummies(df, drop_first=True)
```
